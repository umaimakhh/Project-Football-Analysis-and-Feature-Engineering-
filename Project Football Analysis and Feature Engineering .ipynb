# 4 Feature Engineering with respect to Goal_diff_count
Making a fun â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(x,y):
    return x - y
```

```python
def apply_complex_function(x): return Goal_diff_count(x['GoalsFor'], x['GoalsAgainst'])
df['Goal_diff'] = df.apply(apply_complex_function, axis=1)
```

```python
df.loc[(df['Goal_diff']==df['Goal_diff'].max())]
```

The team with maxium goal is Real Madrid

```python
df.loc[(df['Goal_diff']==df['Goal_diff'].min())]
```

The team with maxium goal is Racing Santander

# 5 Feature Engineering with respect to Winning Percent

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

```python
def winningPercent(X,Y):
    try:
        return (X/Y)*100
    except ValueError:
        return 0
        
```

```python
def applyWinningPercent(x): return winningPercent(x['GamesWon'],astype(int),x['GamesPlayed'].astype(int))
```

```python
#df['winning_Percent']= df.apply(applyWinningPercent)
```

```python
df['GamesWon'].fillna(0,inplace = True)
df['GamesPlayed'].fillna(0,inplace = True)

df['Winning Percent'] = (df['GamesWon'].astype(int)/df['GamesPlayed'].astype(int)) *100

df['Winning Percent'].fillna(0,inplace = True)
```

Printing the top 5 teams which has the highest Winning percentage



```python
df.nlargest(5, 'Winning Percent')
```
